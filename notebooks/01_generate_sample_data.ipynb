```python
import pandas as pd 
import numpy as np 
```

```python
#generating sythetic prediction

np.random.seed(42)

n = 500

confidence = np.random.uniform(0.5, 1.0, n)

true_accuracy = 0.7

correct = np.random.binomial(1, true_accuracy, n)

df = pd.DataFrame({
    "confidence": confidence,
    "correct": correct
})

df.head()
```

```python
df["confidence"] = df["confidence"] + 0.15
df["confidence"] = df["confidence"].clip(0, 1)

df.head()
```

```python
df.to_csv("../sample_data/predictions.csv", index=False)

print("Dataset saved!")
```
